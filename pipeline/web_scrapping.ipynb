```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
```

```python
url          = "https://geo.public.data.uu.nl/vault-globgm/research-globgm%5B1669042611%5D/original/output/version_1.0/transient_1958-2015/"
response     = requests.get(url)
html_content = response.text
soup         = BeautifulSoup(html_content, "html.parser")
```

```python
links = soup.find_all('a')

my_map = {}
for link in links:
    href = link["href"]
    if ".tif" in href and "-bot-" not in href:
        fn, _ = href.split(".")
        _, _, date_str = fn.split("-")
        datetime_object = datetime.strptime(date_str, "%Y%m%d")
        my_map[pd.to_datetime(datetime_object, format="%Y%m")] = href
my_map
```

```python
hrs = pd.date_range(start="1958-01-01 00:00:00", end="1958-01-01 02:00:00", freq="h")
hrs
```

```python
for hr in hrs:
    t = hr.to_period("M").to_timestamp(how="end")
    print(t)
```

```python
result = set(my_map[hr.to_period("M").to_timestamp(how="end").normalize()] for hr in hrs)
print(result)
```
